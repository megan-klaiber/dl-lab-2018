```python
import numpy as np
import json
import os
import matplotlib.pyplot as plt
from collections import OrderedDict
from operator import *
```

### Visualize runs

### cartpole

```python
# import json files rewards
cartpole_path = 'cartpole/'

cartpole_files = ['reward_episode_training.json', 'reward_evaluation_training.json']
cartpole_data = {}

for file in cartpole_files:
    with open(os.path.join(cartpole_path + file)) as json_file:
        cartpole_data[file] = json.load(json_file)
        
cartpole_data['reward_episode_training.json'] = [e[1:3] for e in cartpole_data['reward_episode_training.json']]
#print(cartpole_data)
```

```python
# reward_episode_training

x = [e[0] for e in cartpole_data['reward_episode_training.json']]
y = [e[1] for e in cartpole_data['reward_episode_training.json']]

plt.plot(x, y)

plt.grid(True)
plt.xlabel("episode")
plt.ylabel("reward")
plt.title("Reward during training episodes")


plt.savefig('report/plots/cartpole_episode_reward.png' , dpi=150)
plt.show()

# reward_evaluation_training

x = [20 * e for e in range(1,len(cartpole_data['reward_evaluation_training.json'])+1)]
y = cartpole_data['reward_evaluation_training.json']
#print(x)

plt.plot(x, y)

plt.grid(True)
plt.xlabel("episode")
plt.ylabel("reward")
plt.title("Evaluation reward during training")


plt.savefig('report/plots/cartpole_evaluation_reward.png' , dpi=150)
plt.show()


```
