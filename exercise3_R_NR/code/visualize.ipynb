```python
import numpy as np
import json
import os
import matplotlib.pyplot as plt
from collections import OrderedDict
from operator import *
```

### Visualize runs

```python
# import json files with validation error
path_runs = 'results_run/results_run_'

files = ['0.json', '1.json', '3.json', '5.json', 'new_model.json', 'with_uniform_sampling.json', 'without_uniform_sampling.json',
        '128.json', '32.json']
paths = [path_runs]
data = {}

#for path in paths:
#    temp = []
for file in files:
    with open(os.path.join(path_runs + file)) as json_file:
        data[file] = json.load(json_file)

#print(data)
```

```python
for test in data:
    name = test[:-5]
    
    n_minibatches = data[test]['n_minibatches']
    epochs = np.arange(1, n_minibatches+1, 100)
    
    valid_error = []
    train_error = []

    for e in epochs:
        
        valid_error.append(data[test]["valid_error"][e])
        train_error.append(data[test]['train_error'][e])
    
    plt.plot(epochs, train_error, label = 'train')
    plt.plot(epochs, valid_error, label = 'valid')

    plt.grid(True)
    plt.xlabel("minibatches")
    plt.ylabel("error")
    plt.legend()
    plt.title("Validation and train Error %s" % str(name) )


    plt.savefig('report/plots/valid_error_%s.png' % str(name) , dpi=150)
    plt.show()
```
